```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('~/Desktop/SampleProblem[3].txt') #Loading data into a dataframe for easier manipulation
df.columns=['Participants', 'Pre', 'Post'] #Changes column names to shorter/easier titles
df 
```

```python
#Now lets look for outliers using boxplots
fig = plt.figure(figsize=(26,10))
width = .3
df_box = df.boxplot(['Pre', 'Post'], return_type='axes', fontsize=18, vert=False)
df_box.set_xlabel('Weight(kgs)', fontsize=18)
df_box.set_xlim(-20,120)
df_box.set_title('Boxplots of Pre and Post weights', fontsize=22)
df_box.grid(False)
fig.savefig('hiii.png')
```

```python
df_copy = df #Discovered an outlier at .1 kg in the Post data set. This is definitly an outlier. 
df_copy
df_copy = df_copy.drop([6],axis=0) #Getting rid of pre & post data for Participant 7.
```

```python
df_copy['Participants'] = np.arange(1, 20, 1) 
#Resetting the participant numbers to go from 1-19 as we now only have 19 participants. 
```

```python
df_copy = df_copy.set_index('Participants') #Setting index to participant numbers. 
```

```python
a = df_copy['Pre'] #I want to look at the distrubution of the Pre

fig = plt.figure(figsize=(8,8))
a1 = a.hist()
a1.set_ylim(0,8)
a1.set_title('Distribution of weights (kg) before intervention', fontsize=18)
a1.set_xlabel('Weight in kgs', fontsize=12)
a1.set_ylabel('Number of participants', fontsize=12)
fig.savefig('hi.png') # plotting and saving the distribution of the 'Pre' data to determine if it's normal
```

```python
shap_pre = stats.shapiro(a) #Checking for normality of Pre weight data
print """The t stat is %f and the pvalue is %f for the weights before the intervention."
We accept the null hypothesis that the data is normally distributed"""  % shap_pre 
```

```python
#Now lets do the same for the 'Post' data. First lets look at the distribution
b = df_copy['Post']
fig = plt.figure(figsize=(8,8))
b1 = b.hist()
b1.set_ylim(0,8)
b1.set_title('Distribution of weights (kg) after intervention', fontsize=18)
b1.set_xlabel('Weight in kgs', fontsize=12)
b1.set_ylabel('Number of participants', fontsize=12)
fig.savefig('bye.png') # plotting and saving the distribution of the 'Pre weight' data
```

```python
#so far the post data doesn't look very normally distributed. Lets do a shapiro wilks test
shap_post = stats.shapiro(b) #Checking for normality of Pre weight data
print """The t stat is %f and the pvalue is %f for the weights before the intervention."
We REJECT the null hypothesis that the data is normally distributed."""  % shap_post 
```

```python
#Data is normal. I'm thinking about running a paired ttest, but I need to make sure its assumptions are met. 
##1. The dependent variable is continuous. 2. The independent variable consists of two related groups. 
##3. There should be no outliers. 4. The distribution of differences of weight (Pre weight - Post weight) should be
##approximately normally distributed. I will now check this 4th assumption. 
df_copy['Post - Pre']=df_copy.Post - df_copy.Pre #creating a new column of differences
diff = df_copy['Post - Pre']
diff1 = sns.distplot(diff)
diff1.set_title('hi')
diff1.set_xlabel('hi')
diff1.set_ylabel('hi')
```

```python
shap_diff = stats.shapiro(diff) #Checking for normality of Pre weight data
print """The t stat is %f and the pvalue is %f for the weights before the intervention."
We accept the null hypothesis that the data is normally distributed."""  % shap_diff
```

```python
#The data is approximately normally distrubuted so to compare the pre and post data I will use a two-tailed Student's t-test. 

print "Running a paired T Test"
ttest = stats.ttest_rel(a,b)
```

```python
print """The t stat is %f and the pvalue is %f for the paired T Test."
We ACCEPT the null hypothesis that the mean difference between the Pre and Post group is not different."""  % ttest
```

```python
fig = plt.figure(figsize=(12,8)) #Plotting group mean results
ax = fig.add_subplot(111)
width = .5
means = [a.mean(), b.mean()]
std = [a.std(), b.std()]
ind = np.arange(len(means))
bar_p = plt.bar(ind, means, width, yerr = std, color = 'black', ecolor='red')

ax.set_title('Group weight means (kgs): Pre and Post intervention', fontsize=18)
xTickMarks = ['Pre', 'Post'] #+str(i) for i in range(1,6)]
ax.set_xticks(ind+.3)
ax.set_xlim(-.1)
xtickNames = ax.set_xticklabels(xTickMarks, fontsize=18)
ax.set_ylabel('weight(kgs)', fontsize=18)
fig.savefig('byeee.png')
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)

df_copy['Percent_change'] = ((df_copy['Post'] - df_copy['Pre']) / df_copy['Pre']) *100
z = df_copy['Percent_change'].mean()
h = df_copy['Percent_change'].std()

df_copy.Percent_change.sort_values().plot(kind='bar', color='black')

ax.axhline(y=z, xmin=0, xmax=19, linewidth=.5, color = 'k', linestyle='dashed')
ax.axhline(y=z+h, xmin=0, xmax=19, linewidth=.5, color = 'r', linestyle='dashed')
ax.axhline(y=z-h, xmin=0, xmax=19, linewidth=.5, color = 'r', linestyle='dashed')
ax.annotate('mean', xy=(16, -3.8), xytext=(16,(z-.5)))
ax.axhline(y=0, xmin=0, xmax=19, linewidth=1, color = 'k')
ax.set_title('Individual percentage weight changes', fontsize=20)
ax.set_ylabel('Percent body weight change', fontsize=18)
ax.set_xlabel('', fontsize=18)
xTickMarks = ['Participant'+str(i) for i in np.arange(1,20)]
xtickNames = ax.set_xticklabels(xTickMarks, fontsize=8)
plt.setp(xtickNames, rotation=45, fontsize=10)
plt.show()
fig.savefig('bye.png')
```

```python
fig = plt.figure(figsize=(18,8))
ax = fig.add_subplot(111)
width = .35

mean_a = df_copy['Pre weight']
mean_b = df_copy['Post weight']

ind = np.arange(len(mean_a))

bar_pre = plt.bar(ind, mean_a, width, color = 'black')
bar_post = plt.bar(ind+width, mean_b, width, color = 'red')

ax.set_title('Individual weights (kgs): Pre and Post intervention', fontsize=18)
xTickMarks = ['Participant'+str(i) for i in np.arange(1,20)]
ax.set_xticks(ind)
ax.set_xlim(-width,len(ind)+width)
xtickNames = ax.set_xticklabels(xTickMarks, fontsize=8)
ax.set_ylabel('weight(kgs)', fontsize=18)
plt.setp(xtickNames, rotation=45, fontsize=10)
plt.show()
fig.savefig('butt.png')
```
